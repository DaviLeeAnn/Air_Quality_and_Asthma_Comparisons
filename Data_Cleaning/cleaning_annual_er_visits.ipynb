# Example of thought process going through data cleaning

### 1. Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### 2. Load in the data set and do initial visualization

```python
annual_er_visits = pd.read_csv('../Data/Annual_ER_Visits_For_Asthma.csv')
print(annual_er_visits)
```

### 3. Using `.describe()` shows **Data Comment** and **Unnamed: 7** are entirely NaN, meaning these columns carry no useful information and are candidates for dropping during cleaning.

```python
annual_er_visits.describe()
```

### 4. Now to clearly visualize what columns there are to work with

```python
annual_er_visits.columns
```

### 5. Double checking that the two columns identified earlier are the only ones that need dropped

```python
annual_er_visits.isna()
annual_er_visits.isna().sum()
```

### 6. Proceeding to the cell to drop columns with no values

```python
#Dropping a few columns that I don't need for my analysis
annual_er_visits.drop(columns=['StateFIPS', 'CountyFIPS', 'Data Comment', 'Unnamed: 7'], inplace=True)
print(annual_er_visits)
```

### 7. Singling out the four chosen counties for analysis

```python
#Condensing down to Fayette, Pulaski, Pike, and Warren counties
kentucky_er_visits = annual_er_visits[annual_er_visits['County'].isin(['Fayette', 'Pulaski', 'Pike', 'Warren'])]
print(kentucky_er_visits)
```
